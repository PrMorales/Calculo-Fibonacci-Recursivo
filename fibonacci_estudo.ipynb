```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def fibonacci(n):
    # Nível 1 de indentação (dentro da função)
    if n == 0:
        return 0
    elif n == 1:
        return 1
    else:
        # Nível 2
        termo1 = fibonacci(n - 1)
        termo2 = fibonacci(n - 2)
        res = termo1 + termo2
        return res

# Fora da função (coluna zero)
print(fibonacci(7))
```

```python
# Execução e Teste
n_termo = 7
resultado = fibonacci(n_termo)
print(f"O termo {n_termo} na sequência de Fibonacci é: {resultado}")
```

```python
# Criamos uma lista com os primeiros 10 números de Fibonacci usando a função
sequencia = [fibonacci(i) for i in range(10)]
```

```python
# Platando a cruva de Fibonacci

plt.figure(figsize=(8, 4))
plt.plot(sequencia, marker='o', linestyle='-', color='purple')
plt.title("A Curva de Fibonacci")
plt.xlabel("Posição (n)")
plt.ylabel("Valor do Termo")
plt.grid(True)
plt.show()
```

**OTIMIZAÇÃO DE CÓDIGO E TESTE DE PERFORMANCE**

```python
import time

# Versão 1: Recursiva Simples
def fib_simples(n):
    if n <= 1: return n
    return fib_simples(n-1) + fib_simples(n-2)

# Versão 2: Otimizada
memo = {}
def fib_otimizado(n):
    if n in memo: return memo[n]
    if n <= 1: return n
    memo[n] = fib_otimizado(n-1) + fib_otimizado(n-2)
    return memo[n]

# Teste de tempo
inicio = time.time()
fib_simples(35)
fim = time.time()
print(f"Tempo da versão simples: {fim - inicio:.4f} segundos")

inicio = time.time()
fib_otimizado(35)
fim = time.time()
print(f"Tempo da versão otimizada: {fim - inicio:.4f} segundos")
```

```python
import ColabTurtlePlus.Turtle as turtle
```

```python
def desenhar_arvore(tamanho_galho, t):
    # Caso Base
    if tamanho_galho < 5:
        return

    # Passo Recursivo
    else:
        t.forward(tamanho_galho)
        t.right(20)
        desenhar_arvore(tamanho_galho - 15, t)
        t.left(40)
        desenhar_arvore(tamanho_galho - 15, t)
        t.right(20)
        t.backward(tamanho_galho)

# --- Configuração específica para a tela do Colab ---
turtle.clearscreen()          # Limpa desenhos anteriores
turtle.setup(500, 500)        # Cria uma tela de 500x500 pixels
turtle.bgcolor("gray")       # Fundo cinza

pincel = turtle.Turtle()
pincel.color("lightblue")
pincel.speed(13)       # Velocidade máxima no ColabTurtlePlus
pincel.pensize(3)

# Posiciona a tartaruga na base
pincel.penup()
pincel.goto(0, -200)   # Vai para o centro-baixo da tela
pincel.setheading(90)  # Aponta para cima (90 graus)
pincel.pendown()

# --- Executa a Mágica ---
desenhar_arvore(80, pincel)
```
